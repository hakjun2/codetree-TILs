```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
from scipy.fftpack import ifft
import librosa

# 음성 데이터 불러오기
audio_file = 'classical.00001.wav'
y, sr = librosa.load(audio_file, sr=None)

# 모음 추출 (예시로 첫 1초의 모음만 사용)
start_time = 0
end_time = sr  # 1초
vowel_signal = y[start_time:end_time]

# Cepstrum 계산
_, _, Sxx = spectrogram(vowel_signal, sr)
cepstrum_result = ifft(np.log(Sxx + 1e-10))  # 역 Fourier 변환을 사용한 Cepstrum 계산

# 결과 시각화
plt.figure(figsize=(12, 8))

# 음성 신호
plt.subplot(3, 1, 1)
plt.plot(y)
plt.title('Original Audio Signal')

# 모음 신호
plt.subplot(3, 1, 2)
plt.plot(vowel_signal)
plt.title('Vowel Signal')

# Cepstrum
plt.subplot(3, 1, 3)
plt.plot(np.abs(cepstrum_result))
plt.title('Cepstrum of Vowel Signal')

plt.tight_layout()
plt.show()
```
